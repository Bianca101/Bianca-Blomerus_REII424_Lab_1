```python
with open('messy.csv','r') as in_file, open('organised.csv','w') as out_file:

    for row in in_file:
     
        #FORMATTING
            #Capitalization
        row = row.upper()
            #Irregular Nulls
        row = row.replace(",,",",NULL,")    
            #Whitespace 
        row=row.strip()
            #Invalid characters
      #  row = row.decode("asci","ignore")
        
        out_file.write(row)
        print(row)
```
