# task 7 duplicate data handling


```python
# Some entries are duplicated. All fields of the entry is duplicated thus has no meaning. 
# I have decided to create a new csv file without duplicates calles noduplicate.csv 
# Original file still contains duplicates
```

```python
import csv
fl =open ("duplicate.csv","r") #open input file for read


with open('duplicate.csv','r') as in_file, open('noduplicate.csv','w') as out_file:
    current = set() # fast lookup
    for line in in_file:
        if line in current: continue # do not add duplicate line to new csv file

        current.add(line)
        out_file.write(line)
        print(line)
```
