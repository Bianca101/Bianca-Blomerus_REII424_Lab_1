# Lab 4 : Data Visualization

```python
#----------------------------------------LAB 4-------------------------------------------
# 24888508
# Dataset used: Iris dataset
# Dataset description: CSV file containing the Sepal lenth, Sepal width, Petal length
# and Petal width of three different iris species
#
#----------------------------------------------------------------------------------------

# Import dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

iris = pd.read_csv("iris2.csv")
#print(iris) #View dataset

#----------------------------------------------------------------------------------------
#
# Task 1: Pie Chart
# Used for exploratory data analysis, when asking "HOW MANY?"
#
#
#----------------------------------------------------------------------------------------


#df = pd.DataFrame (ds['data'], columns = ds['feature_names'])
df = pd.DataFrame(iris, columns=['Sepal length','Sepal width','Petal length','Petal width','Species','Time'])
df

group_species = df.groupby('Species').sum()
var = 'Sepal width'
group_species[var].plot(kind='pie', fontsize=20)
plt.ylabel(var, horizontalalignment='left')
plt.title('PIE CHART: Breakdown for ' + var, fontsize=25)
plt.show()
plt.savefig('Pie chart.jpg')
#plt.close()

#----------------------------------------------------------------------------------------
#
# Task 2: Bar chart
# Used for exploratory data analysis as well, and is sometimes a better choice than a pie chart 
# when wanting to visualize relative sizes
#
#----------------------------------------------------------------------------------------


group_species[var].plot(kind='bar', fontsize=15,
rot=30)
plt.title('Breakdown for ' + var, fontsize=20)
plt.savefig('Bar Chart.jpg')
plt.close()
group_species = df.groupby('Species').sum()
group_species.plot(
 kind='bar', subplots=True, fontsize=12)
plt.suptitle('Total Measurements, by Species')
plt.savefig('Bar chart multi.jpg')
#plt.close()


#----------------------------------------------------------------------------------------
#
# Task 3: Histogram
# Used for determining outliers and variation in data
#
#
#----------------------------------------------------------------------------------------

df.plot(kind="hist", subplots=True)#, layout=(2,2))
plt.suptitle('Iris Histograms', fontsize=20)
plt.savefig('Histogram.jpg')
plt.show()
#plt.close()
for spec in df['Species'].unique():
    forspec = df[df['Species']==spec]
    forspec['Petal length'].plot(kind='hist', alpha=0.4, label=spec)
    plt.legend(loc='upper right')
plt.suptitle('Petal Length by Species')
#plt.close()

#----------------------------------------------------------------------------------------
#
# Task 4: Boxplot
# Used for summarizing data, show min, max, mean and quantiles
#
#
#----------------------------------------------------------------------------------------

col = 'Sepal length'
df['ind'] = pd.Series(df.index).apply(lambda i: i% 50)
df.pivot('ind','Species')[col].plot(kind='box')
plt.savefig('Boxplot.jpg')
#plt.close()

#----------------------------------------------------------------------------------------
#
# Task 5: Scatterplot
# Used to visualize relationships within datasets
#
#
#----------------------------------------------------------------------------------------

df.plot(kind="scatter", x="Sepal length", y="Sepal width")
plt.title("Sepal Length vs Sepal Width")
plt.savefig("Scatterplot")
#plt.close()


#----------------------------------------------------------------------------------------
#
# Task 6: Heatmap
# Used to reduce clutter in scatterplots if there is a lot of data,
# can distinguish between areas with fewer or more points
#
#----------------------------------------------------------------------------------------

df.plot(kind="hexbin", x="Sepal width", y="Sepal length")
plt.title("Sepal length vs Sepal width")
plt.savefig("Heatmap")
#plt.close()

#----------------------------------------------------------------------------------------
#
# Task 7: Time Series
# Very important data type with no substitute for plotting. Any data with time related attributes
# can be visualized with a time series plot
#
#----------------------------------------------------------------------------------------

df.plot(x="Time", y ="Sepal length")
plt.title("Random Time series Plot")
plt.ylabel(" ")
#plt.close()

```
